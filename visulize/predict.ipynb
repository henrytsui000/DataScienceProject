```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('dark_background')
```

```python
price = pd.read_csv("../data/predict_value.csv", index_col=0)
```

```python
companys = dict()
for company in price.columns[:5]:
    companys[company] = (price[company], price[company+"_pred"])
```

```python
price = price.iloc[-200:-100]
```

```python
price_len = len(price.index)
step = 5
x_tick = range(0, price_len, price_len//step)
```

```python
pic = sns.lineplot(data = price, dashes=False)
pic.set_xticks(x_tick, labels=price.index[x_tick])
plt.savefig("../src/predict/test.png", transparent = True)
```
